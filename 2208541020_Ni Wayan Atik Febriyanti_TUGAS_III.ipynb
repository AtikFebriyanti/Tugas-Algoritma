```python
###########################################
#String Methods ada beberapa metode yaitu :
###########################################

#####################################################################################
# 1. capitalize yang berfungsi untuk mengonversi karakter pertama menjadi huruf besar
#####################################################################################
judul ='bali merupakan pulau seribu pura'
print (judul.capitalize())
```

```python
#################################################################################
# 8.format berfungsi untuk memberikan format nilai yang ditentukan didalam string.
#################################################################################
text1 ="Nama saya adalah {nama}, umur saya {umur}".format(nama = 'Atik' ,umur =18)
text2 ="Nama saya adalah {0}, umur saya {1}".format('Atik',18)
text3 ="Nama saya adalah {}, umur saya {}".format('Atik',18)
print (text1)
print (text2)
print (text3)
```

```python
#################################################################################################
# operasi matematika diphyton adalah kontruksi yang dapat memanipulasi nilai dari operan 
# sebagai contoh 1 + 2 = 3 ,dimana 1 & 2 adalah operan dan + adalah operator,ada beberapa yaitu :
#################################################################################################

#########################################################################################################################
# 1. penjumlahan berfungsi untuk menjumlahkan nilai dari masing-masing operan atau bilangan.
# 2. pengurangan berfungsi untuk mengurangi nilai operan disebelah kiri menggunakan operan sebelah kanan.
# 3. perkalian berfungsi untuk mengalikan operan atau bilangan. 
# 4. pembagian berfungsi untuk membagi operan disebelah kiri menggunakan operan disebelah kanan.
# 5. sisa bagi berfungsi untuk mendapatkan sisa pembagian dari operan sebelah kiri yang dibagi oleh operan sebelah kanan.
# 6. pangkat berfungsi untuk memangkatkan operan sebelah kiri dengan operan sebelah kanan operator.
# 7. pembagian bulat memiliki fungsi yang sama seperti pembagian ,yang membedakan hanya angka dibelakang koma dihilangkan.
#########################################################################################################################
print(7+6)
print(25-17)
print(17*47)
print(2856/2)
print(87%4)
print(67**2)
print(87//4)
print(5+67*4-65)
```

```python
#####################################################################################################
# operator perbandingan di phyton adalah operator yang bertugas untuk membandingkan antara dua operan.
# apabila hasil perbandingan benar maka akan menghasilkan nilai TRUE. 
# sebaliknya apabila hasil perbandingan salah maka akan menghasilkan nilai FALSE.
#####################################################################################################
x=10>5
print(x)
x=5<10
print(x)
x=5>10
print(x)
x=10>=5
print(x)
x=5<=10
print(x)
x=10 == 10
print (x)
x=10 == 5
print (x)
x=10 !=4
print (x)
```

```python
#############################################################################################################################
# operator logika di phyton adalah operator yang digunakan untuk membuat kesimpulan logis dari 2 kondisi yaitu TRUE dan FALSE.
# beberapa operator yaitu and ,or dan not.
# operator and akan bernilai TRUE jika kedua nilai adalah TRUE.
# operator or akan bernilai TRUE jika salah satu nilainya adalah TRUE. 
# operator not akan menghasilkan nilai yang berlawanan.
#############################################################################################################################
buku_tulis = 20
buku_cerita = 15
print(buku_tulis > 10 and buku_cerita < 20 )
```

```python
print(buku_tulis < 10 or buku_cerita <20)
print(buku_tulis<10 or buku_cerita > 20)
```

```python
print(not (buku_tulis < 10 or buku_cerita > 20))  
```

```python
##########################################################################################
# operator if di phyton akan dijalankan apabila nilai variabel yang dijalankan adalah TRUE.
# if adalah kode program yang dijalakan apabila bernilai TRUE.
# else adalah kode program yang dijalankan apabila bernilai FALSE. 
##########################################################################################
x = 35
if( x % 3) == 0:
  print( 'variabel x berisi angka genap')
if( x % 3) != 0:
  print('variabel x berisi angka ganjil')
```
