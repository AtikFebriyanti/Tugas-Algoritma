5.11 WHILE LOOP

```python
# while loop merupakan metode perulangan dimana ada kondisi yang harus dipenuhi agar looping bisa berjalan terus.
# apabila syarat while sebelum ":" sudah terpenuhi maka variabel akan masuk kefungsi.
# untuk mengulang suatu proses yang belum diketahui jumlahnya maka akan dilakukan pengecekan ,jika kondisi masih bernilai true makan looping terus berlanjut.

x = 0
while -5 <= x <= 4:
  y = x*2
  x = x-1
  print('dari nilai x =',( x*2 + x-2),'adalah nilai dari y yang memenuhi syarat')
  print( x,' yang merupakan nilai dari x yang memenuhi syarat ')
```

5.13 METODE LIST

5.14 FOR LOOP

```python
# for loop merupakan statement control flow untuk iterasi khusus yang memungkinkan kodeuntuk dijalankan berulangan.
# for loop berguna untuk menjalankan perintah yang perlu dilakukan secara berulang-ulang.
# perbedaan dengan while adalah kalau while perulangan yang memiliki syarat ,sedangkan for perulangan tanpa syarat.

makanan = ['betutu','ayam bakar','bakso','nasi goreng']
for jenis in makanan:
  print(jenis)
print('1.' +makanan[0] + ' khas gilimanuk yang rasanya enak sekali.')
print('2.' +makanan[1] + ' sangat enak apabila dimakan sama nasi hangat ,sambal matah,dan plecing kangkung.')
print('3.' +makanan[2] + ' adalah makanan yang sangat enak dimakan kapan saja ,apalagi pada saat lagi hujan.')
print('4.' +makanan[3] + ' akan lebih nikmat apabila ditambahkan dengan krupuk dan acar timun.')
print('5.' +makanan[0] +','+ makanan[1] +','+ makanan[2] + ' dan'+ makanan[3] + ' merupakan makan khas dari negara INDONESIA')
```

```python
z = 0
while z <len(makanan):
  print(makanan[z])
  z+=1
```

```python
for z in range(len(makanan)):
  print(makanan[z])
```

5.15 FUNGSI RANGE 

```python
# fungsi ini berguna untuk memberikan bilangan berurutan atau berpola sesuai dengan argumen yang di inginkan.
# kita bisa menggunkan range untuk menyederhanakan penulisan perulangan for.
# saat input satu,range (X) artinya dari 0 hingga x
# saat input dua,range (X,Y) artinya dari X hingga Y

angka = range (8)
print(angka)
```

```python
for isi in angka:
  print(isi)
```

```python
for isi in range (5,15):
  print(isi)
```

```python
for isi in range (0,80,8):
  print(isi)
```
